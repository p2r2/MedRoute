```python
import time

def run_referral_model(input_text):
    time.sleep(1) 
    return f"GENERATED REFERRAL: based on '{input_text[:20]}...'"

def run_ehr_model(input_text):
    print("\n>>> 🗂️ ROUTED TO: EHR/Clinical Data Model")
    print("    [System] Connecting to secure patient database context...")
    time.sleep(1) 
    # logic to call actual model would go here
    return f"EHR QUERY RESULT: Found patient data related to '{input_text[:20]}...'"

def run_billing_model(input_text):
    print("\n>>> 💰 ROUTED TO: Billing & Finance Model")
    print("    [System] Accessing CPT/ICD-10 coding database...")
    time.sleep(1) 
    # logic to call actual model would go here
    return f"BILLING CODE: Calculated costs for '{input_text[:20]}...'"

def run_general_fallback(input_text):
    print("\n>>> 🤖 ROUTED TO: General Assistant (Fallback)")
    return "I am not sure which specialist to use. I will handle this as a general query."

# --- 2. The Rule-Based Router Logic ---


# --- 3. Main Loop (Input from You) ---

def main():

    while True:
        try:
            # Get input from user typing
            user_query = input("Enter your medical query: ")
            
            if user_query.lower() in ['exit', 'quit']:
                print("Shutting down...")
                break
            
            if not user_query.strip():
                continue

           
            result = route_request(user_query)
            
            
            print(f"    [Output] {result}\n")
            print("-" * 50)

        except KeyboardInterrupt:
            print("\nShutting down...")
            break

if __name__ == "__main__":
    main()
```

```python
#llm routing logic- rule based

def route_request(user_input):
    
    text = user_input.lower()
    
    billing_keywords = ['invoice', 'cost', 'price', 'insurance', 'copay', 'deductible', 'bill', 'code', 'cpt']
    if any(word in text for word in billing_keywords):
        return run_billing_model(user_input)

    
    referral_keywords = ['refer', 'letter', 'specialist', 'handover', 'transfer', 'recommendation']
    if any(word in text for word in referral_keywords):
        return run_referral_model(user_input)


    ehr_keywords = ['history', 'lab', 'result', 'blood', 'prescription', 'medication', 'dose', 'patient id', 'mrn']
    if any(word in text for word in ehr_keywords):
        return run_ehr_model(user_input)

    
    return run_general_fallback(user_input)
```

```python
import os
from openai import OpenAI
import json

def run_billing_model(user_query, api_key=None):

    
    secure_key = api_key or os.getenv("MY_AI_API_KEY")
    
    if not secure_key:
        return "[Error]: API Key is missing."

    client = OpenAI(api_key=secure_key)

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "Extract the 'patient_name' and 'target_data' from the query. Return ONLY JSON."},
                {"role": "user", "content": user_query}
            ]
        )
        
        ai_extraction = response.choices[0].message.content
        
    
        data_json = json.loads(ai_extraction)
        name = data_json.get('patient_name', 'Unknown')
        
        return f"SUCCESS: Found records for Patient [{name}].\n    (Simulated DB Output): Last Vitals: Normal, Lab ID: #99281"

    except Exception as e:
        return f"[Error in EHR Model]: {str(e)}"
```
